```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from docx import Document
import PyPDF2
```

```python
def read_resume(file_path):
    text = ""
    if file_path.endswith(".pdf"):
        with open(file_path,'rb') as f:
            reader = PyPDF2.PdfReader(f)
            for page in reader.pages:
                text += page.extract_text() + ""
    elif file_path.endswith(".docx"):
        doc = Document(file_path)
        for p in doc.paragraphs:
            text += p.text + ""
            return text.lower()
```

```python
def read_id(file_path):
    with open(file_path,"r",encoding="utf-8")as f:
        return f.read().lower()
```

```python
def ats_score(resume_text,jd_text):
    vectorizer = TfidfVectorizer(stop_words = 'english')
    vectors = vectorizer.fit_transform([resume_text,jd_text])
    score = cosine_similarity(vectors[0:1],vectors[1:2])[0][0]
    return round(score * 100,2)
```

```python
skills_db = pd.read_csv(r"E:\AI_Lab\datasets\skills_master.csv")
def missing_skills(resume_text,jd_text):
    jd_skills = []
    for skills in skills_db['skills']:
        if skill in jd_text:
            jd_skills.append(skill)

    resume_skills = [skill for skill in jd_skills if skill in resume_text]
    missing = list(set(jd_skills)-set(resume_skills))
    return missing
```

```python
def improvement_tips(missing):
    tips = []
    for m in missing:
        tips.append(f"Add '(m)' skill with 1-2 real project bullet points.")
        return tips
```

```python
def read_jd(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read().lower()
```

```python
def read_resume(path):
    with open(path, "r", encoding="utf-8") as f:
        return f.read().lower()
```

```python
resume = read_resume(r"E:\AI_LAB\datasets\resume.txt")
jd = read_jd(r"E:\AI_LAB\datasets\job_description.txt")

score = ats_score(resume, jd)
missing = missing_skills(resume, jd)
tips = improvement_tips(missing)

print("ATS SCORE:", score, "%")
print("Missing Skills:", missing)
print("Suggestions:")
for t in tips:
    print("-", t)
```
